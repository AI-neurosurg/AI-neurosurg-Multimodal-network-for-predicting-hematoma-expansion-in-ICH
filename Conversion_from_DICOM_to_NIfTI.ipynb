```python
import os
from tqdm import tqdm
import dicom2nifti
import dicom2nifti.settings as settings
```

## Function

```python
def dicom_to_nifti_gantry_tilted(dicom_dir, nifti_dir, name):

    """
    Parameters::  dicom_dir: the path of the directory where the DICOM data to be converted is stored
                  nifti_dir: the path of the directory where the converted nifti files are stored
                  name: the name to add to the file name for file identification
    """

    settings.disable_validate_orthogonal()
    settings.enable_resampling()
    settings.set_resample_spline_interpolation_order(1)
    settings.set_resample_padding(-1000)

    dicom_dir_subdir = []
    for dicomdir, subdirs, files in os.walk(dicom_dir):
        for names in subdirs:
            dicom_dir_subdir.append(os.path.join(dicom_dir, names))
    dicom_dir_subdir.sort()
    dicom_dir_subdir
    for i, subdir in enumerate(tqdm(dicom_dir_subdir)):
        nifti1 = dicom2nifti.dicom_series_to_nifti(
            subdir, os.path.join(nifti_dir, name + os.path.basename(subdir) + '.nii.gz'))
    return nifti1
```

## Specify the directories for DICOM and NIfTI files

```python
dicom_dir = "PATH"
nifti_dir = "PATH"
name = "HOSPITAL NAME"
```

## Conversion

```python
dicom_to_nifti_gantry_tilted(dicom_dir, nifti_dir, name)
```
